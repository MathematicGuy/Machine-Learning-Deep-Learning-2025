### Mutable & Immutable objects (Đối tượng có thể và không thể thay đổi)
+ Mutable types like `list, dict, set` can be altered in place. 

+ Immutable types like `int, str, tuple` cannot be change in place; any "change" creates a new object. 



```python
x = 10
y = [1, 2, 3]

def func(a, b):
	a = a + 5 #? changing a (immutable) to new object
	b.append(4) #? changing a mutable object modified the object
	
	print("ínide:", a, b)
	
func(x, y)
print('outside:', x, y)
```

### Inner Scope & Outer Scope

```python
def outer():
	x = 5
	def inner():
		x = 10
		print("Inner x:", x)
	inner()
	print("Outer x:", x)
outer()
```
