# Q5 - UNet Skip Connection: Concat vs Add
Notebook này dùng cho câu hỏi Q5 về so sánh `concat` và `add` trong skip connection của decoder.

Thiết lập của bài:
- Tensor sau up-conv có shape `[1, 32, 32, 32]`.
- Tensor skip từ encoder có shape `[1, 32, 32, 32]`.
- Hai tensor đã cùng số kênh, nên có thể dùng cả `concat` lẫn `add`.
- Các lớp convolution đều dùng `bias=False`.

Mục tiêu:
1. Hoàn thiện block decoder.
2. Chạy forward cho cả hai chế độ `concat` và `add`.
3. Quan sát số tham số của `conv1` để chọn đáp án đúng.

```python
import torch
import torch.nn as nn

torch.manual_seed(42)
```

```python
class DecoderBlock(nn.Module):
    def __init__(self, merge_mode='concat'):
        super().__init__()
        self.merge_mode = merge_mode
        self.up = nn.ConvTranspose2d(64, 32, kernel_size=2, stride=2, bias=False)

        if merge_mode == 'concat':
            in_channels = None  # TODO: điền số kênh input cho conv1 khi dùng concat
        elif merge_mode == 'add':
            in_channels = None  # TODO: điền số kênh input cho conv1 khi dùng add
        else:
            raise ValueError("merge_mode must be 'concat' or 'add'")

        self.conv1 = nn.Conv2d(in_channels, 32, kernel_size=3, padding=1, bias=False)

    def forward(self, x, skip):
        x = self.up(x)

        if self.merge_mode == 'concat':
            # TODO: ghép skip và x theo chiều kênh
            raise NotImplementedError('TODO: implement concat merge')
        else:
            # TODO: cộng từng phần tử giữa skip và x
            raise NotImplementedError('TODO: implement add merge')

        return self.conv1(x)
```

```python
x = torch.randn(1, 64, 16, 16)
skip = torch.randn(1, 32, 32, 32)

block_concat = DecoderBlock(merge_mode='concat')
block_add = DecoderBlock(merge_mode='add')

y_concat = block_concat(x, skip)
y_add = block_add(x, skip)

print('Output shape - concat:', y_concat.shape)
print('Output shape - add   :', y_add.shape)
print('numel:', block_concat.conv1.weight.numel(), block_add.conv1.weight.numel())
```

## Gợi ý
- `concat` sẽ làm thay đổi số kênh sau merge.
- `add` là phép cộng element-wise nên số kênh sau merge giữ nguyên.
- Bạn có thể in `x.shape` sau bước merge để tự kiểm tra.
