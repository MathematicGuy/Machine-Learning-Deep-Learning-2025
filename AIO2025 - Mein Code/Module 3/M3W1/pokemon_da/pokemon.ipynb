```python
import pandas as pd

# Create a simple pandas Series
example_series = pd.Series([10, 20, 30, 40, 50])
print(example_series)

def increase_hp(row):
    return row * 5

# Using .map to multiply each element by 2
mapped_series = example_series.map(lambda x: x * 5)
print("Mapped Series:")
print(mapped_series)

# Using .apply to multiply each element by 2
applied_series = example_series.apply(increase_hp)
print("\nApplied Series:")
print(applied_series)
```

### NA Filling technique

Method to use for filling holes in reindexed Series:
+ ffill: propagate last valid observation forward to next valid.
+ backfill / bfill: use next valid observation to fill gap.


```python
import numpy as np

#? .fillna() is deprecated, use ffill or bfill instead
test_df = pd.Series([1,3, np.nan, 4, np.nan, np.nan, 5])

#? fill all NA with the last valid value until the next valid forward/backward)
print(test_df.copy().ffill()) # use the last/final valid observation to fill all missing values from that point, until the next valid valiue
print('mean:', test_df.mean())
print(test_df.copy().bfill()) # use the last/final valid observation to fill all missing values from that point, until the next valid valiue
```
